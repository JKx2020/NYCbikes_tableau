```python
import pandas as pd 
from matplotlib import pyplot as plt 
import numpy as np 
from scipy import stats
import os.path
import sys
from datetime import datetime as dt
```

```python
#read in the tripdata from citibike & union together
dec17 = "../sourceData/tripdata/JC-201712-citibike-tripdata.csv"
dec17_df = pd.read_csv(dec17)
dec17_df['filename'] = os.path.basename(dec17)

jan18 = "../sourceData/tripdata/JC-201801-citibike-tripdata.csv"
jan18_df = pd.read_csv(jan18)
jan18_df['filename'] = os.path.basename(jan18)

# feb18 = "../sourceData/tripdata/JC-201802-citibike-tripdata.csv"
# feb18_df = pd.read_csv(feb18)
# feb18_df['filename'] = os.path.basename(feb18)

# mar18 = "../sourceData/tripdata/JC-201803-citibike-tripdata.csv"
# mar18_df = pd.read_csv(mar18)
# mar18_df['filename'] = os.path.basename(mar18)

apr18 = "../sourceData/tripdata/JC-201804-citibike-tripdata.csv"
apr18_df = pd.read_csv(apr18)
apr18_df['filename'] = os.path.basename(apr18)

# may18 = "../sourceData/tripdata/JC-201805-citibike-tripdata.csv"
# may18_df = pd.read_csv(may18)
# may18_df['filename'] = os.path.basename(may18)

# jun18 = "../sourceData/tripdata/JC-201806-citibike-tripdata.csv"
# jun18_df = pd.read_csv(jun18)
# jun18_df['filename'] = os.path.basename(jun18)

jul18 = "../sourceData/tripdata/JC-201807-citibike-tripdata.csv"
jul18_df = pd.read_csv(jul18)
jul18_df['filename'] = os.path.basename(jul18)

#union the dfs together
# bike_data = pd.concat([jan18_df, feb18_df, mar18_df, apr18_df, may18_df, jun18_df])

# bike_data = pd.concat([jan18_df, feb18_df, mar18_df])

bike_data = pd.concat([dec17_df, jan18_df, apr18_df, jul18_df])

#reset the index
bike_data = bike_data.reset_index(drop=True)

# bike_data = jan18_df.reset_index(drop=True)

bike_data.head()
```

```python
#get all the columns in the csv
bike_data.columns
```

```python
#update the M F code from 0,1,2
bike_data['gender'] = bike_data['gender'].replace({0:'unknown',1: 'male',2:'female'})
bike_data.head()
```

```python
#get the number of month and create a column
bike_data['month'] = bike_data.filename.str[7:9]
bike_data['year'] = bike_data.filename.str[3:7]
bike_data['trip_id'] = bike_data.index
bike_data['path_id'] = bike_data["start station name"] + "-" + bike_data["end station name"]

bike_data.head()
```

```python
#create an origin df, get rid of the end station data
bike_data["origin-destination"] = "origin"
origin_df = bike_data.drop("end station id", axis=1)
origin_df = origin_df.drop("end station name", axis=1)
origin_df = origin_df.drop("end station latitude", axis=1)
origin_df = origin_df.drop("end station longitude", axis=1)
origin_df.head()
```

```python
#create a destination df, get rid of the origin station data
bike_data["origin-destination"] = "destination"
destination_df = bike_data.drop("start station id", axis=1)
destination_df = destination_df.drop("start station name", axis=1)
destination_df = destination_df.drop("start station latitude", axis=1)
destination_df = destination_df.drop("start station longitude", axis=1)
destination_df
destination_df.head()
```

```python
#rename the columns in the two new dataframes to get rid of the start and end in the columns headers
destination_df = destination_df.rename(columns={"end station id":"station id", "end station name": "station name", "end station latitude":"station latitude","end station longitude": "station longitude"})

origin_df = origin_df.rename(columns={"start station id":"station id", "start station name": "station name", "start station latitude":"station latitude","start station longitude": "station longitude"})

origin_df.head()
```

```python
#union the DFs together
bike_origin_destination_df = pd.concat([origin_df, destination_df]).sort_values(by=["trip_id"])
bike_origin_destination_df.head()
```

```python
#add in the day of the week to the df
# bike_origin_destination_df["day_of_week"] = pd.to_datetime(bike_origin_destination_df["starttime"].astype('int'))

bike_origin_destination_df["day_of_week"] = pd.to_datetime(bike_origin_destination_df["starttime"]).dt.strftime('%w').astype('int')


bike_origin_destination_df.head()
```

```python
bike_origin_destination_df.to_csv("../cleanedData/bike_origin_destination_df.csv")
bike_data.to_csv("../cleanedData/clean_bike_data.csv")
```
